```python
cars = sns.load_dataset('mpg') # Import the cars dataset from seaborn
cars
```

```python
manufacturer = cars['name'].str.split(expand=True)[0] #alternatively .iloc[:,0] 
cars['Manufacturer']= manufacturer

cars['name'].str.replace('', '-')
cars
```

```python
# find the average weight for usa
#select weight for usa only
weightusa = cars.loc[cars['origin']== 'usa', 'weight']
# calculate the average
Ave_weightusa = weightusa.mean()

#select weight for japan only
weightjapan = cars.loc[cars['origin']== 'japan', 'weight']
# calculate the average
Ave_weightjapan = weightjapan.mean()

#select weight for usa only
weighteurope = cars.loc[cars['origin']== 'europe', 'weight']
# calculate the average
Ave_weighteurope = weighteurope.mean()

print(" The total weight of USA, Japan, and Europe are {0:.2f}, {0:.2f}, and,{0:.2f}, respectively". format( Ave_weightusa ,Ave_weightjapan, Ave_weighteurope) ) 
```

```python
sales = pd.read_csv(r"C:\Users\Ramatu's PC\Desktop\430pm\ABC Sales.csv")
sales
```

- Create three new columns showing total revenue, total cost and total profit for each transaction. 

```python
sales['Revenue'] = sales['Quantiy Sold']*sales['Unit Sale Price ($)']
sales['Cost'] = sales['Quantiy Sold']*sales['Unit Cost Price ($)']
sales['Profit'] = sales.Revenue - sales.Cost
sales
```

- Sort the resultant dataframe based on total profit column in descending order? Which product transaction is most profitable?  

```python
#sort the dataframe by profit in descending order
sales.sort_values(by='Profit', ascending = False) 
sales
```

- Use groupby and appropriate aggregation method to find the total revenue this store is generating from online and In store sales. 

```python
Total_Rev = sales.groupby(['Order Channel'])['Revenue'].sum()
Total_Rev
```
