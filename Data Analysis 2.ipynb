# Part 2: Indexing, Slicing and Subsetting DataFrames

##### Simple Indexing
- Selecting data using column name
- Slicing subsets of rows
- Subsetting using boolean indexing


### 1. Simple Indexing: Selecting data using column name


```python
# Select the time column
df['time'] # or df.time
```

```python
#select the time and sex column
df[['time', 'sex']]
```

```python
#from the previos subset select sex column
df[['time', 'sex']]['sex']
```

### 2. Simple Indexing: Slicing subsets of rows

```python
# select the first 3 rows
df[0:3]
```

```python
# select the first 5 rows
df[:5]
```

```python
# select the 15th to 25th rows
df[15:26]
```

```python
#select the last five rows
df[-5:]
```

```python
#select the last row
df[-1:]
```

### 3. Simple Indexing: Subsetting using boolean indexing

```python
# Outputs true or false
df['sex'] == 'Male'
```

```python
# The same as above but filters the Table for only "True" cases
df[df.sex == 'Male']
```

```python
# Select rows where tips is greater than or equal to 5
df[df.tip >= 5]
```

```python
# select only the total bill and tips columns for rows where the tip is greater than 5
df[df.tip > 5][['total_bill', 'tip']]
```

```python
#select the sex and smoker columns for case where smoker is False
df[df['smoker'] != 'Yes'][['sex', 'smoker']]
```

##### Loc Indexer Method - explicit index and Boolean indexer
- Sub-setting by index names
- Sub-setting by Boolean indexing




### 4. Loc - Sub-setting by index names

```python
# df.loc[rows,cols]
# Select the first 10 rows of the sex column
df.loc[0:10, 'sex']
```

```python
#select all the rows for the columns total bill and sex
df.loc[:,['total_bill','sex']]
```

```python
# Select the first 10 rows of the sex and tip column
df.loc[0:10, ['sex','tip']]
```

```python
#select the first 5 rows of all the columns
df.loc[:5,:]
```

```python
#Rows up to 3,Columns up to smoker
df.loc[:3,:'smoker']
```

### 5. Loc - subsetting by Boolean indexing


```python
# Select the sex and tip colums for rows where smoker = Yes
df.loc[df['smoker'] == 'Yes', ['sex', 'tip']]
```

```python
#show the total bill,time and size for dinner only
df.loc[df['time']=='Dinner',['total_bill','time','size']]
```

```python
#show the total bill,time and size for sunday only
df.loc[df['day']=='Sun',['total_bill','time','size']]
```

```python
# if all the condition holds
df.loc[(df['smoker'] == 'Yes')& (df['day'] == 'Sun') & (df['size']>3),:]
```

```python
#if any of the condition holds
df.loc[(df['smoker'] == 'Yes') | (df['day'] == 'Sun') | (df['size']>3),:]
```

```python
#create new column
# On average how much do male smokers spend in the restaurant for their food bill (excluding tips)?
df['Bill Without Tip'] = df['total_bill'] - df['tip']
Average_billMS = df.loc[(df['sex']=='Male') & (df['smoker'] == 'Yes'), 'Bill Without Tip'].mean()
print("On Average Male Smokers spend {0:.2f} (excluding tip) for their food bill".format(Average_billMS))
```

##### iLoc Indexer Method
- Sub-setting by index position


### 6. iLoc: Sub-setting by index position

```python
#select the 10th to 19th row and the fourth colum
df.iloc[10:20, 4]
```

```python
#select the 10th to 19th rows for the 3rd to 6th column
df.iloc[10:20, 3:7]
```

```python
#select the second row and the 6th column
df.iloc[2, 6]
```

### 7. Seaborn
see documentation: https://seaborn.pydata.org/tutorial.html

- visualizing Statistical relationship

```python
#Scather plots
# x and y axis numerical
sns.scatterplot(x='total_bill', y = 'tip', data = df, hue = 'smoker') #shift tab
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", style="smoker",
            data=df);
```

- Visualizing distributions of data

```python
sns.distplot(df['size'], bins=[1, 2, 3, 4, 5, 6, 7])
```

- Plotting with categorical data


```python
sns.catplot(x='total_bill', data = df, kind = 'box')
```

```python
sns.catplot(x='day', y = 'total_bill', kind = 'bar', data = df) 
#default is mean, estimator = np.max, est can be equal to any aggregate value
```

```python
sns.catplot(x="smoker", y="tip", order=["No", "Yes"], data=df)
```

```python
sns.catplot(x="day", y="total_bill", hue="smoker", kind="box", data=df)
```

```python
df["weekend"] = df["day"].isin(["Sat", "Sun"])
sns.catplot(x="day", y="total_bill", hue="weekend",
            kind="box", dodge=False, data=df)
```

### 8. Matplotlib

- Visualizing Statistical relationship

```python
colors = {'Yes':'blue', 'No':'orange'}
plt.scatter('total_bill', 'tip', c = df['smoker'].map(colors), data=df)
plt.xlabel('Total Bill')
plt.ylabel('Tip')

plt.show()
```

- Visualizing distributions of data

```python
plt.hist(df['size'], bins = [1, 2, 3, 4, 5, 6, 7], density=True, facecolor='g', alpha=0.75)

plt.xlabel('Size')
plt.ylabel('Frequency')
plt.title('Histogram of Size')
plt.show()
```

- Plotting with categorical data


```python
plt.boxplot(df['total_bill'],patch_artist=True,vert=False)
plt.xlabel('Total Bill')
plt.show()
```

```python
plt.bar(df['day'], df['total_bill'])
plt.xlabel('Total Bill')
plt.show()
```
